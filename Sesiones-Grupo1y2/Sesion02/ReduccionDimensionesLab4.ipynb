# Graficación PairPlot
Trazar relaciones por pares en un conjunto de datos.

Por defecto, esta función creará una cuadrícula de Ejes de manera que cada variable data se compartirá en el eje y en una sola fila y en el eje x en una sola columna. Los ejes diagonales se tratan de manera diferente, dibujando un gráfico para mostrar la distribución univariada de los datos para la variable en esa columna.

También es posible mostrar un subconjunto de variables o trazar diferentes variables en las filas y columnas.

Esta es una interfaz de alto nivel para la PairGridque es fácil dibujar algunos estilos comunes. Debe usarlo PairGrid directamente si necesita más flexibilidad.

Dibuje diagramas de dispersión para relaciones conjuntas e histogramas para distribuciones univariadas:

```python
# import required libraries
import seaborn as sns; sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
```

Muestra diferentes niveles de una variable categórica por el color de los elementos de la trama:

```python
 g = sns.pairplot(iris, hue="species")
```

Use una paleta de colores diferente:

```python
g = sns.pairplot(iris, hue="species", palette="husl")
```

Use marcadores diferentes para cada nivel de la variable de tono:

```python
g = sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

Trazar un subconjunto de variables:

```python
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

Se puede definir el tamaño de la gráfica:

```python
g = sns.pairplot(iris, height=3,
...                  vars=["sepal_width", "sepal_length"])
```

Utilizando  diferentes variables en las filas y columnas:

```python
g = sns.pairplot(iris,
...                  x_vars=["sepal_width", "sepal_length"],
...                  y_vars=["petal_width", "petal_length"])
```

Utilizando estimaciones de densidad de kernel  para graficaciones univariadas:

```python
g = sns.pairplot(iris, diag_kind="kde")
```

Ajustando modelos de regresión lineal a los diagramas de dispersión:

```python
g = sns.pairplot(iris, kind="reg")
```

Pase argumentos de palabras clave a las funciones subyacentes (puede ser más fácil de usar PairGriddirectamente):

```python
g = sns.pairplot(iris, diag_kind="kde", markers="+",
...                  plot_kws=dict(s=50, edgecolor="b", linewidth=1),
...                  diag_kws=dict(shade=True))
```
